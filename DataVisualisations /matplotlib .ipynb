```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

#### **`Problem-1:`** Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China. 

```python
# Loading DataSet
df = pd.read_csv('https://tinyurl.com/2fe6vz4u')
df.head()
```

```python
iran = df[df['Country'] == 'China'].groupby(by='Year')['PM2.5'].sum()
china = df[df['Country'] == 'Iran'].groupby(by='Year')['PM2.5'].sum()
china
```

```python
plt.plot(iran.index, iran.values, label = 'Iran', linestyle= 'dashed', marker= 'o')
plt.plot(china.index, china.values, label = 'China', marker = 'o')
plt.xlabel("Year")
plt.ylabel('PM2.5 over years')
plt.xticks(df['Year'].value_counts().index)
plt.legend()
plt.grid()
plt.show()
```

#### **`Problem-2:`** Draw a histogram of the  column "PM10" of which the y-axis represents the probability (see the documentation how to draw the probability). 

```python
df['PM10'].value_counts().index
```

```python
plt.hist(df['PM10'], density=True, bins=50, facecolor='green')
plt.xlabel('Bins')
plt.ylabel('Probability')
plt.title('Histogram of PM10')
plt.grid()
plt.show()
```

#### **`Problem-3:`** Draw a scatter plot where x-axis represents "PM2.5" and y-axis represents "PM10" for two countries Poland and Chile.

```python
poland = df[df['Country'] == 'Poland']
chile = df[df['Country'] == 'Chile']
```

```python
plt.scatter(x = poland['PM2.5'], y = poland['PM10'], label = 'Poland')
plt.scatter(x = chile['PM2.5'], y = chile['PM10'], label = 'Chile')
plt.xlabel('PM2.5')
plt.ylabel('PM10')
plt.title('PM2.5 vs PM10 for Poland and Chile')
plt.legend()
plt.grid()
plt.show()
```

#### `Problem-4:` Draw a pie chart of top 5 most frequent countries.

```python
freq = df['Country'].value_counts()[:5]
explode = (0.1, 0, 0, 0, 0)
plt.pie(freq, autopct = '%1.1f%%', labels = freq.index, explode = explode, startangle=90)
plt.show()
```

#### **`Problem-5:`** Draw a bar chart which represents the counts of top 5 most frequent countries.



```python
plt.bar(freq.index, freq.values)
plt.xlabel('Country')
plt.ylabel('Frequency Count')
plt.show()
```
